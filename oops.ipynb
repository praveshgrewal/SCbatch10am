### class and object

```python
class Car:
    def __init__(self,color,brand,model,engine):
        self.color = color
        self.brand = brand
        self.model = model
        self.engine = engine
    
    def full_details(self):
        return f"{self.color}:{self.brand}:{self.model}:{self.engine}"


# object
my_car = Car('black','BMW','x7','Petrol')
print(my_car.full_details())
```

```python
sahil_car = Car('white','creata')
print(sahil_car.brand)
print(sahil_car.color)
```

```python
class Car:
    def __init__(self,color,brand):
        self.color = color
        self.brand = brand
    
    def poly(self):
        return "this is car class"


class toy_car(Car):
    def __init__(self,color,brand,price):
        super().__init__(color,brand)
        self.price = price
    
    def poly(self):
        return "this is toy car class"


my_car = Car('black','tata')
my_car_toy = toy_car('black','tata','12L')
print(my_car.poly())
print(my_car_toy.poly())
```
