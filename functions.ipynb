### parameter and argumetns

```python
def greet(name,age): # parameter
    return f"good morning {name}{age}"

print(greet('piyush',45)) # argument
```

```python
def is_even(n):
    if n%2==0:
        return "even"
    else:
        return "odd"

print(is_even(23))
```

```python
def is_evens(n):
    return n%2==0

print(is_evens(22))
```

```python
def calc(a,b):
    return a+b,a-b,a*b

add,sub,multi = calc(12,8)
print(add)
print(sub)
print(multi)
```

### default argumetn

```python
def greet(name="user"): # default argument
    return f"good morning {name}"

print(greet('sahil')) 
```

### *args
* muliple arguments

```python
def greet(*args):
    return f"good morning {args}"

print(greet('sahil','piyush','ankit',23)) 
```

### ** kwargs
* if you want argumetns in dict form

```python
def greet(**args):
    return f"good morning {args}"

print(greet(name= "sahil",age=45,address = "delhi")) 
```

### local variable and global variabe

```python
def local_var():
    global name12
    name12 = "rizwan" # local variable
    return name12

print(local_var())
```
