### parameter and argumetns

```python
def greet(name,age): # parameter
    return f"good morning {name}{age}"

print(greet('piyush',45)) # argument
```

```python
def is_even(n):
    if n%2==0:
        return "even"
    else:
        return "odd"

print(is_even(23))
```

```python
def is_evens(n):
    return n%2==0

print(is_evens(22))
```

```python
def calc(a,b):
    return a+b,a-b,a*b

add,sub,multi = calc(12,8)
print(add)
print(sub)
print(multi)
```

### default argumetn

```python
def greet(name="user"): # default argument
    return f"good morning {name}"

print(greet('sahil')) 
```

### *args
* muliple arguments

```python
def greet(*args):
    return f"good morning {args}"

print(greet('sahil','piyush','ankit',23)) 
```

### ** kwargs
* if you want argumetns in dict form

```python
def greet(**args):
    return f"good morning {args}"

print(greet(name= "sahil",age=45,address = "delhi")) 
```

### local variable and global variabe

## lambda function

```python
def squ(num):
    return num* num

print(squ(3))
```

```python
square = lambda x:x*x
print(square(3))
```

```python
adds = lambda a,b,c,d,e:a+b+c+d+e
print(adds(12,13,16,78,89))
```

```python
nums = [1,2,4,6,9,3]
squ_nums = list(map(lambda x:x*x,nums))
print(squ_nums)
```

```python
names = ['rizwan','asish','ayan','sahil','danish','deepika']
upper_names = list(map(lambda x :x.upper(),names))
print(upper_names)

```

```python
nums = [1,2,4,6,9,3,78,23,78,22,90,223]
even_nums = list(filter(lambda x:x%2 == 0,nums))
print(even_nums)
```

```python
nums = [1,2,4,6,9,3,78,23,78,22,90,223]
even_nums = list(filter(lambda x:x>=40,nums))
print(even_nums)
```

## recursive function

```python
def factorial(n):
    if n ==0 or n ==1:
        return 1
    else:
        return n * factorial(n-1)

factorial(5)
```
