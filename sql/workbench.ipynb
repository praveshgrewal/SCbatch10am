```python
# connection

testing = mysql.connector.connect(
  host="localhost",
  user="root",
  password="sudoroot"
)

print(testing)
print("connected successfully")
```

```python
cursor.execute(""" create database datascience """)
print("database created")
```

```python
cursor.execute("""

CREATE TABLE datascience.Students (
    id INT PRIMARY KEY,
    name VARCHAR(50) NOT NULL,
    location VARCHAR(50),
    grade char(50),
    marks int
);


""")

print("table created")
```
