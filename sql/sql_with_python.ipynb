```python
import sqlite3
```

```python
# data base create (or connect)

connect = sqlite3.connect('test.db')
print('database created')
```

```python
# connection (cursor ,sql query execute)
cursor = connect.cursor()
```

```python
# table create
cursor.execute("""

CREATE TABLE Students (
    id INT PRIMARY KEY,
    name VARCHAR(50) NOT NULL,
    location VARCHAR(50),
    grade char(50),
    marks int
);


""")

print("table created")
```
