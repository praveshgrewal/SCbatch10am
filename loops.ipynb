## jumping statments
* break ---> if you want to stop execution
* continue ---> skip the current iteration
* pass ----> as palceholder

```python
for i in range(1,10):
    if i == 5:
        continue
    print(i)
```

```python
name = "danish ur rehman"
for char in name:
    if char == " ":
        break

    print(char)
```
