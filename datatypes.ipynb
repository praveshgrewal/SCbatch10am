# data types

## int ---> number 

```python
age = 22
print(age)
print(type(age)) # type ---> check the datatype of variable or value
```

```python
type(12)
```

## string(str) ---> text --- " ",' ', """ """, ''' '''

```python
name = """ ahmed """
print(name)
print(type(name))
```

```python
# my name is "khan"
```

```python
print('my name is "khan" ')
```

## float ---> decimal values

```python
score = 86.7
print(score)
print(type(score))
```

## boolean --- >True/False

```python
t1 = True  # --> 1
print(t1)
print(type(t1))
```

```python
t2 = False # --->0
print(t2)
print(type(t2))
```

## NoneType -- >nothing(null)

```python
no = None
print(no)
print(type(no))
```

## complex number ---> real+img

```python
com = 3+5j
print(com)
print(type(com))
```

## list ---> [ ]

```python
names = ['ankit',23,'ashish',22,'rizwan',23,'piyush',20,'deepika',21,'danish',22,'sahil',22]
print(names)
print(type(names))
```

## tuple ---> ( )

```python
names2= ('ankit',23,'ashish',22,'rizwan',23,'piyush',20,'deepika',21,'danish',22,'sahil',22)
print(names2)
print(type(names2))
```

## set --- { }

```python
names3= {'ankit',23,'ashish',22,'rizwan',23,'piyush',20,'deepika',21,'danish',22,'sahil',22}
print(names3)
print(type(names3))
```

## dictionary ---> dict ---> {'key':'value'}

```python
my_dict = {'name':'jiger','age':34}
print(my_dict)
print(type(my_dict))
```

# type casting / type conversion

```python
age5 = "ankit"
print(type(age5))
print(age5)
age6 = int(age5) # convert age5 into int
print(type(age6))
print(age6)
```

```python
num2 = 10
print(type(num2))
num3 = 56.9
print(type(num3))
output = num2+num3
print(type(output))
```
