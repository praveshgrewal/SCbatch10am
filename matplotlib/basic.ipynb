### bar plot

```python
names = ['deepika','asish','ankit']
marks = [34,89,72]


colors = ['purple','green','blue']
plt.bar(names,marks,color=colors,width=0.6,edgecolor='red',linewidth=2)
plt.title('marks of students')
plt.show()
```

```python
bars = plt.bar(names, marks)
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             y,
             y,
             ha='center',
             va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Laptop', 'Mobile', 'Tablet', 'Monitor']
sales = [50, 80, 30, 60]

plt.figure(figsize=(8,5))

bars = plt.bar(categories, sales,
        color=['blue','orange','green','red'],
        edgecolor='black')

plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Units Sold")

plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,
             y+1,
             y,
             ha='center')

plt.show()
```

## scatter plot

```python
x = [1,2,3,4,5,6,9,8,7,4,6,3,8,5,2,0,1]

y = [10,20,30,40,50,8,7,10,69,85,40,78,6,78,45,88,78]
plt.scatter(x,y,marker='o',s = 100, alpha=0.9,edgecolor='black',c=x, cmap='Blues')
plt.grid()
plt.colorbar()

plt.show()
```

```python
import matplotlib.pyplot as plt

hours = [1,2,3,4,5,6,7]
marks = [50,55,60,65,70,80,90]

plt.figure(figsize=(8,5))

plt.scatter(hours,
            marks,
            color='blue',
            s=120,
            marker='o',
            edgecolor='black',
            alpha=0.7)

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = [1,2,3,4,5]
y = [5,6,7,8,9]
z = [10,15,20,25,30]

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.scatter3D(x, y, z)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, color='red')

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

ax.set_title("3D Scatter Plot")

plt.show()
```

```python
import numpy as np

x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)

x,y = np.meshgrid(x,y)

z = x**2 + y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x,y,z)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='plasma')
ax.set_title('3D Surface Plot')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

##  pie chart

```python
course = ['DS','DA','DM','SDE','Cyber']
data = [20,25,15,30,10]
colors = ['#4E79A7', '#F28E2B', '#E15759', '#76B7B2', '#59A14F']
explode = (0,0,0,0.1,0)
plt.figure(figsize=(9,6))
plt.axis('equal')
plt.pie(data,labels=course,colors=colors,explode=explode,
        shadow=True,
        autopct='%1.0f%%')
plt.legend(course,loc='upper right')
plt.show()
```
