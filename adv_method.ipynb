## zip 

```python
names = ['rizwan','ashish','deepika','aiman danish ur rehman','sahil']
ages = [23,22,22,23,22]

for n,a in zip(names,ages):
    print(n,a)
```

```python
fav_car = ['fortuner','breeza','scorpio','ertica','ferrai f1','mercdes','range rover']

for i,f in enumerate(fav_car,start=1):
    print(f"{i}:{f}")
```
