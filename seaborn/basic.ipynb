```python
import seaborn as sns
import matplotlib.pyplot as plt

# set theme
sns.set_theme(style="whitegrid")

# set figure size
plt.figure(figsize=(10,6))

# create bar plot
sns.barplot(
    data=tips,
    x='sex',
    y='tip',
    hue='time',
    palette='viridis',
    edgecolor='black',
    linewidth=1.5
)

# add title
plt.title("Average Tip Amount by Gender and Time", fontsize=16, fontweight='bold')

# axis labels
plt.xlabel("Customer Gender", fontsize=12)
plt.ylabel("Average Tip ($)", fontsize=12)

# grid customization
plt.grid(axis='y', linestyle='--', alpha=0.6)

# legend customization
plt.legend(title="Meal Time", loc='upper left')

# adjust layout
plt.tight_layout()

# show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# set professional theme
sns.set_theme(style="whitegrid", font_scale=1.2)

# create pairplot
g = sns.pairplot(
    data=iris,
    hue='species',
    palette='husl',
    markers=['o','s','D'],
    diag_kind='kde',
    plot_kws={
        'alpha':0.7,
        's':70,
        'edgecolor':'black'
    }
)

# add title
g.fig.suptitle("Pairwise Relationships in Iris Dataset", fontsize=18, fontweight='bold')

# adjust spacing
g.fig.tight_layout()

# fix title overlap
g.fig.subplots_adjust(top=0.95)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# set theme
sns.set_theme(style="whitegrid")

# set figure size
plt.figure(figsize=(8,5))

# create count plot
ax = sns.countplot(
    data=tips,
    x='sex',
    palette='Set2',
    edgecolor='black',
    linewidth=1.5
)

# add title
plt.title("Number of Customers by Gender", fontsize=16, fontweight='bold')

# axis labels
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Customer Count", fontsize=12)

# add grid only on y-axis
plt.grid(axis='y', linestyle='--', alpha=0.6)

# add value labels on bars
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',
        (p.get_x() + p.get_width()/2., p.get_height()),
        ha='center',
        va='bottom',
        fontsize=11
    )

# adjust layout
plt.tight_layout()

plt.show()
```

### heatmap

```python
import pandas as pd

data = {
    'Math':[85,78,92,70,88],
    'Science':[90,74,89,65,91],
    'English':[75,80,85,60,87],
    'History':[88,76,90,72,84]
}

df = pd.DataFrame(data,
                  index=['Student1','Student2','Student3','Student4','Student5'])

df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.heatmap(df.corr(),
            annot=True,
            cmap='coolwarm',
            linewidths=1)

plt.title("Student Marks Heatmap")
plt.xlabel("Subjects")
plt.ylabel("Students")

plt.show()
```
