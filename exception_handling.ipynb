```python
try:
    num = int(input('Enter Your Number: '))
    num2 = int(input('Enter Your 2nd Number: '))
    result = num/num2
    print(result)
except ZeroDivisionError:
    print("can't divide any value from zero")
except TypeError:
    print("check value again")
except ValueError:
    print("check value again")
finally:
    print("execution completed")

```
